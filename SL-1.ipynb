# Data pre-processing

```python
#split dependent and independent variables:
from imblearn.over_sampling import SMOTE
from collections import Counter
from matplotlib import pyplot
x=data[["P_incidence","P_tilt","L_angle","S_slope","P_radius","S_Degree"]]
y=data["Class"]
oversample = SMOTE()
X, Y = oversample.fit_resample(x, y)
print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(Y))
```

```python
from sklearn import preprocessing
standardized_X = preprocessing.scale(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(standardized_X,Y,test_size=0.3,random_state=7)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

myList = list(range(3,20))

# subsetting just the odd ones
neighbors = list(filter(lambda x: x % 2 != 0, myList))

ac_scores = []

# perform accuracy metrics for values from 1,3,5....19
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    # predict the response
    y_pred = knn.predict(x_test)
    # evaluate accuracy
    scores = accuracy_score(y_test, y_pred)
    print("Accuracy score for {} is {}".format(k,scores))
    ac_scores.append(scores)

# changing to misclassification error
MSE = [1 - x for x in ac_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print("The optimal number of neighbors is %d" % optimal_k)
```

K=3 have good accuracy when compared to the rest.

```python
import matplotlib.pyplot as plt
# plot misclassification error vs k
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```

# Applying tuning techniques

1. Parameter tuning using cross_val_score

```python
from sklearn.model_selection import cross_val_score
k_range = range(3, 31)

# list of scores from k_range
k_scores = []

# 1. we will loop through reasonable values of k
for Kval in k_range:
    # 2. run KNeighborsClassifier with k neighbours
    knn = KNeighborsClassifier(n_neighbors=Kval)
    # 3. obtain cross_val_score for KNeighborsClassifier with k neighbours
    scores = cross_val_score(knn, x_test, y_test, cv=15, scoring='accuracy')
    # 4. append mean of scores for k neighbors to k_scores list
    k_scores.append(scores.mean())
print(k_scores)
```

On applying the cross validation technique our optimal K value 3 score is reduced to 0.80 from 0.89

2. Applying Grid searchCV

```python
from sklearn.model_selection import GridSearchCV
k_range = list(range(3, 31))
grid_scores=[]
param_grid = dict(n_neighbors=k_range)
grid = GridSearchCV(knn, param_grid, cv=15, scoring='accuracy')
grid.fit(x_test, y_test)
grid.cv_results_
```

On applying the Search grid CV technique our optimal K value 3 score is reduced to 0.80 from 0.89

3. Applying Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV

k_range = list(range(3, 31))

param_dist = dict(n_neighbors=k_range)

rand = RandomizedSearchCV(knn, param_dist, cv=15, scoring='accuracy', n_iter=10, random_state=5)

rand.fit(x_test, y_test)

rand.cv_results_
```

On applying the Randomized CV technique our optimal K value 3 score is reduced to 0.80 from 0.90
